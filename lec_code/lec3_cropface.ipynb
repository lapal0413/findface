```python
import numpy as np
from numpy.linalg import norm
```

```python
def cosine_sim(A,B):
    cosine = np.dot(A,B)/(norm(A)*norm(B))
    return cosine
```

```python
cosine_sim(embedding_target,embedding_0)
```
