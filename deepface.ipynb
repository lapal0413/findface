```python
from deepface import DeepFace
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
backends = [
  'opencv', 
  'ssd', 
  'dlib', 
  'mtcnn', 
  'retinaface', 
  'mediapipe',
  'yolov8',
  'yunet',
  'fastmtcnn',
]
```

```python
image = cv.imread("face.jpg")
```

```python
cvt_image=cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(cvt_image)
```

```python
face_objs = DeepFace.extract_faces(img_path = "face.jpg",         target_size = (224, 224),         detector_backend = backends[4]) 
```

```python
pt1=(face_objs[0]["facial_area"]["x"],face_objs[0]["facial_area"]["y"])
pt2=(face_objs[0]["facial_area"]["x"]+face_objs[0]["facial_area"]["w"],face_objs[0]["facial_area"]["y"]+face_objs[0]["facial_area"]["h"])
color=(255,0,0)
box_image = cv.rectangle(cvt_image, pt1, pt2, color,5) 
```

```python
plt.imshow(box_image)
```
